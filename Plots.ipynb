```python
import matplotlib.pyplot as plt
import numpy as np

# Data
letters = ['A', 'E', 'I', 'N', 'L', 'R', 'O', 'T', 'U', 'S', 'M', 'C', 'H', 'D', 'B', 'G', 'Y', 'P', 'V', 'F', 'J', 'K']
values = [3500000, 3200000, 2900000, 2600000, 2300000, 2000000, 1900000, 1800000, 1700000, 1600000, 1500000, 1400000, 
          1300000, 1200000, 1100000, 1000000, 900000, 800000, 700000, 600000, 500000, 400000]

# Plot
fig, ax = plt.subplots()
bars = ax.bar(letters, values, color='#ADD8E6')

# Add value labels
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 50000, f'{yval:,}', ha='center', va='bottom')

# Set labels and title
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_title('')

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Remove ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add gridlines
ax.yaxis.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
names = ['THOMAS', 'EMMA', 'LUCAS', 'CAMILLE', 'HUGO', 'SATYNE', 'CHATALIC', 'HEMME', 'TOULIER ANCIAN', 'TRAMBLE']
values1 = [2271, 1870, 1851, 1722, 1715]
values2 = [1, 1, 1, 1, 1]

# Combine values1 and values2 to match names list
values = values1 + values2

x = np.arange(len(names))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
bars1 = ax.bar(x[:5] - width/2, values1, width, label='Group 1', color='#ADD8E6')
bars2 = ax.bar(x[5:] + width/2, values2, width, label='Group 2', color='#FF69B4')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_title('')
ax.set_xticks(x)
ax.set_xticklabels(names, rotation=45, ha='right')
ax.legend()

# Add labels above bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(bars1)
autolabel(bars2)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Remove ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add gridlines
ax.yaxis.grid(True)

fig.tight_layout()

plt.show()
```
